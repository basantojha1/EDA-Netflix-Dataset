# DATA CLEANING PROCESS

```python
# Checking for missing number of values
df.isnull().sum()
```

```python
# checking the percentage of missing values
df.isnull().mean()*100
```

```python
# Handeling the missing values in director as shown from the data above
df[df['director'].isnull()]
```

```python
# Populating the null values with unknown
df['director'] = df['director'].fillna('unknown')
```

```python
df['director'].isnull().sum()
```

```python
# Handeling missing values in cast
df[df['cast'].isnull()]
```

```python
# Populating the null values with unknown
df['cast'] = df['cast'].fillna('unknown')
df['cast'].isnull().sum()
```

```python
# Handeling Missing values in country
df[df['country'].isnull()]
```

```python
# Populating the null values with unknown
df['country'] = df['country'].fillna('unknown')
df['country'].isnull().sum()
```

```python
# Since date_added, rating and duration have less than 10 null values, we can directly drop that row
df.dropna(inplace = True)
```

```python
# Checking again to confirm all missing values are dealth with
df.isnull().sum()
```

```python
# Checking for Data Mode
df.info()
```

```python
df.head()
```

```python
# Since the Data is in correct formate we can move on to Standarization
```

# Standarization

```python
df.country.head()
```

```python
# Capitalizing the country using lambda function
df['country'] =df['country'].apply(lambda x: x.capitalize())
df.country.head()
```

```python
df[['date_added', 'release_year']]
```

```python
# Splitting the date_added to year and month
df['added_month'] = df['date_added'].apply(lambda x: x.split(',')[0].split()[0])
df['added_year'] = df['date_added'].apply(lambda x: x.split(',')[1])
```

```python
df[['added_month', 'added_year']]
```

```python
df
```

```python
# dropping "date_added" column
df.drop('date_added', axis=1, inplace=True)
```

```python
# Data type conversion
df['added_year'] = df['added_year'].astype(int)
```

```python
df[df['added_year']<df['release_year']]
```

```python
# since added year cannot be less than release year, we will be droping those rows.
df.drop(df[df['added_year']<df['release_year']].index, inplace = True)
```
